```python
import tensorflow as tf
import image_dataset_utils as utils
import os

os.environ["KMP_DUPLICATE_LIB_OK"] = "True" # Prevents OpenMP runtime error when showing images

tf.get_logger().setLevel('ERROR') # Suppress TensorFlow warnings like "no registered converter for this op"

print("TensorFlow version:", tf.__version__)
```

```python
loaded_model = tf.keras.models.load_model("car_classifier_model.h5")
loaded_class_names = utils.load_class_names_from_file("augmented_cars/class_names.txt")

print("Model loaded successfully.")
```

```python
test_ces, _ = utils.load_one_image("cars/ces1.jpg")
predictions = loaded_model.predict(test_ces)
predicted_class = tf.argmax(predictions[0]).numpy()

predicted_class_name = loaded_class_names[predicted_class]
utils.show_image(test_ces, "Predicted class: " + predicted_class_name)
```
