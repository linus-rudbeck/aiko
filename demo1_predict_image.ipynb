```python
import tensorflow as tf
import image_dataset_utils as utils
import os
import requests

os.environ["KMP_DUPLICATE_LIB_OK"] = "True"

print("TensorFlow version:", tf.__version__)
```

```python
image_path = "data/test/dog-artan.jpeg"
preprocessed_image, _ = utils.load_one_image(image_path)
utils.show_image(preprocessed_image, "Ärtan: Golden Retriever")
```

```python
model = tf.keras.applications.MobileNetV2(input_shape=(128,128,3),
                                               include_top=True,
                                               weights='imagenet')

print("Model loaded successfully.")
```

```python
predictions = model(preprocessed_image)
predicted_class = tf.argmax(predictions, axis=-1).numpy()[0]
print(f"Predicted class: {predicted_class}")
```

```python
labels_path = "https://raw.githubusercontent.com/pytorch/hub/master/imagenet_classes.txt"
classes = requests.get(labels_path).text.splitlines()
classes.insert(0, "") # Add empty string for index 0

print(f"Predicted class: {classes[predicted_class]}")
```
